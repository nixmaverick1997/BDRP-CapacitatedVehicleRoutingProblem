```python
import jax
import jumanji
```

Instantiating Jumanji environment

```python
#CVRP problem with 20 nodes (randomly generated), maximum capacity of 30, and maximum demand of 10.
from jumanji.environments.combinatorial.cvrp.env import CVRP
from jumanji.environments.combinatorial.cvrp.types import Observation, State
from jumanji.wrappers import AutoResetWrapper
```

```python
env=CVRP()
env=AutoResetWrapper(env)
```

```python
key = jax.random.PRNGKey(0)
```

```python
print(key)
```

```python
state, timestep = jax.jit(env.reset)(key)
```

```python
print(state, timestep)
```

```python
action = env.action_spec().generate_value()
```

```python
state, timestep = jax.jit(env.step)(state, action)  
```

```python
print(state, timestep)
```
